### Sélection des Représentants

```python
x = X.flatten().tolist()

nb_clusters = 2
ndim = 2

native_clusters = r.init_clusters(x, nb_clusters, ndim)

clusters = r.npify(native_clusters, nb_clusters*ndim)
print(clusters)

plt.scatter(clusters[0], clusters[1], color='lightgrey', s=1000)
plt.scatter(clusters[2], clusters[3], color='lightgrey', s=1000)

r.k_means(x, native_clusters, nb_clusters, ndim)

clusters = r.npify(native_clusters, nb_clusters*ndim)
print(clusters)

r.destroy(native_clusters, nb_clusters*ndim)

plt.scatter(clusters[0], clusters[1], color='lightgreen', s=400)
plt.scatter(clusters[2], clusters[3], color='lightgreen', s=400)
plt.scatter(X[0, 0], X[0, 1], color='blue', s=50)
plt.scatter(X[1:3,0], X[1:3,1], color='red', s=50)
plt.show()
plt.clf()
```

### Entrainement du Modèle

```python
x = X.flatten().tolist()
y = np.array([1.0, -1.0])
clusters = np.array([1.0, 1.0, 2.5, 3.0])


ndim = 2
gamma = 2.5
model_size = int(len(clusters) / ndim)
model = r.create_linear_model(model_size-1)
print(r.npify(model, model_size))

points_x1_blue = []
points_x2_blue = []
points_x1_red = []
points_x2_red = []


for _ in range(5):
    points_x1_blue = []
    points_x2_blue = []
    points_x1_red = []
    points_x2_red = []
    for i in np.arange(-0.0, 4.0, 0.1):
        for j in np.arange(-0.0, 4.0, 0.1):
            predict = r.predict_rbf_naive(model, clusters, np.array([i,j]), ndim, gamma, True)
#             print(f"{_} {predict_reg} {i} {j}")
            if predict >= 0.0:
                points_x1_blue.append(i)
                points_x2_blue.append(j)
            else:
                points_x1_red.append(i)
                points_x2_red.append(j)

    r.train_rosenblatt_rbf(model, clusters, y, ndim, 20, 0.01, 0.1)
#     print(r.npify(model, model_size))

plt.scatter(points_x1_red, points_x2_red, c=color_2)
plt.scatter(points_x1_blue, points_x2_blue, c=color_1)
plt.scatter(clusters[0], clusters[1], color='lightgreen', s=300)
plt.scatter(clusters[2], clusters[3], color='lightgreen', s=300)
plt.scatter(X[0, 0], X[0, 1], color='blue', s=50)
plt.scatter(X[1:3,0], X[1:3,1], color='red', s=50)
plt.show()

r.destroy(model, model_size)
```
