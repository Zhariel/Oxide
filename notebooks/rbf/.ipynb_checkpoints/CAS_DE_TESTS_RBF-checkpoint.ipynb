# Imports

```python
import matplotlib.pyplot as plt
import numpy as np
from ruster import Ruster
import time
import sys
import os

color_1 = "salmon"
color_2 = "deepskyblue"
color_3 = "lightgreen"

r = Ruster(path="C:\\Users\\Revive\\PycharmProjects\\Rust_Native_ML\\rust\\PerceptronLib\\target\\release\\PerceptronLib.dll")

print(r.sum([1, 2, 3]))
```

## Classification

### Linear Simple :
        Linear Model : OK
        MLP (2, 1)   : OK

```python
X = np.array([
      [1, 1],
      [2, 3],
      [3, 3]
])
Y = np.array([
      1,
      -1,
      -1
])
```

```python
plt.scatter(X[0, 0], X[0, 1], color='blue', s=50)
plt.scatter(X[1:3,0], X[1:3,1], color='red', s=50)
plt.show()
plt.clf()
```

### RBF avec Repr√©sentants

```python
x = X.flatten().tolist()

nb_clusters = 2
ndim = 2

native_clusters = r.init_clusters(x, nb_clusters, ndim)

clusters = r.npify(native_clusters, nb_clusters*ndim)
print(clusters)

plt.scatter(clusters[0], clusters[1], color='lightgrey', s=1000)
plt.scatter(clusters[2], clusters[3], color='lightgrey', s=1000)

r.k_means(x, native_clusters, nb_clusters, ndim)

clusters = r.npify(native_clusters, nb_clusters*ndim)
print(clusters)

r.destroy(native_clusters, nb_clusters*ndim)

plt.scatter(clusters[0], clusters[1], color='lightgreen', s=400)
plt.scatter(clusters[2], clusters[3], color='lightgreen', s=400)
plt.scatter(X[0, 0], X[0, 1], color='blue', s=50)
plt.scatter(X[1:3,0], X[1:3,1], color='red', s=50)
plt.show()
plt.clf()
```

```python
clusters = np.array([1.0,  1.0,  2.5, 3.0 ])
Y = np.array([
    1,
    -1
])

model_size = 2
native_model = r.create_linear_model(model_size)
model = r.npify(native_model, model_size+1)
print(model)

points_x1_blue = []
points_x2_blue = []
points_x1_red = []
points_x2_red = []

for _ in range(20):
    points_x1_blue = []
    points_x2_blue = []
    points_x1_red = []
    points_x2_red = []
    for i in np.arange(1.0, 3.1, 0.07):
        for j in np.arange(1.0, 3.1, 0.07):
            predict = r.predict_linear_model_classification(native_model, np.array([i, j], float), (model_size+1))
            if predict == 1.0:
                points_x1_blue.append(i)
                points_x2_blue.append(j)
            elif predict == -1.0:
                points_x1_red.append(i)
                points_x2_red.append(j)
    
    r.train_rosenblatt_linear_model(native_model, clusters, Y, model_size+1, 20, 0.01)
    model = r.npify(native_model, model_size+1)

plt.scatter(points_x1_blue, points_x2_blue, c=color_1)
plt.scatter(points_x1_red, points_x2_red, c=color_2)
plt.scatter(clusters[0], clusters[1], color='lightgreen', s=400)
plt.scatter(clusters[2], clusters[3], color='lightgreen', s=400)
plt.scatter(X[0, 0], X[0, 1], color='blue', s=50)
plt.scatter(X[1:3,0], X[1:3,1], color='red', s=50)
plt.show()

print(_)
print(clusters)
print(model)

r.destroy(native_model, model_size)
```

### Multi Linear 3 classes :
        Linear Model x3 : OK
        MLP (2, 3)      : OK

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([[1, 0, 0] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 1, 0] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else 
              [0, 0, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else 
              [0, 0, 0]for p in X])
X = X[[not np.all(arr == [0, 0, 0]) for arr in Y]]
Y = Y[[not np.all(arr == [0, 0, 0]) for arr in Y]]
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,1], color='red')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,1], color='green')
plt.show()
plt.clf()
```
